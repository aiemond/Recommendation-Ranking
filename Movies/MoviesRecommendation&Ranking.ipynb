```python
# Calculate all the components based on the above formula
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movie_sorted_ranking=movies_cleaned_df.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort movies by weighted average in descending order
sorted_movies = movie_sorted_ranking.sort_values(by='weighted_average', ascending=False)

# Create a dot plot of the top 10 movies based on weighted average
plt.figure(figsize=(12, 6))
top_movies = sorted_movies.head(10)
sns.stripplot(x=top_movies['weighted_average'], y=top_movies['original_title'], data=top_movies, jitter=True, palette='Set2')

# Set plot limits and labels
plt.xlim(4, 10)
plt.title('Top 10 Movies by Weighted Average Votes (Dot Plot)', fontweight='bold')
plt.xlabel('Weighted Average Score', fontweight='bold')
plt.ylabel('Movie Title', fontweight='bold')

# Save the plot as an image file
plt.savefig('top_movies_dotplot.png')

```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply the scaler to the 'weighted_average' and 'popularity' columns
scaled_features = scaler.fit_transform(movies_cleaned_df[['weighted_average', 'popularity']])

# Create a new DataFrame with the scaled values and appropriate column names
normalized_movies_df = pd.DataFrame(scaled_features, columns=['weighted_average', 'popularity'])

# Display the first few rows of the normalized DataFrame
normalized_movies_df.head()
```

```python
# Add the normalized columns to the original DataFrame
movies_cleaned_df[['normalized_weight_average', 'normalized_popularity']] = normalized_movies_df

# Display the first few rows of the updated DataFrame to verify
movies_cleaned_df.head()
```

```python
# Compute the score as a weighted average of normalized columns
movies_cleaned_df['score'] = (movies_cleaned_df['normalized_weight_average'] * 0.5 +
                              movies_cleaned_df['normalized_popularity'] * 0.5)

# Sort the DataFrame by the computed score in descending order
movies_scored_df = movies_cleaned_df.sort_values(by='score', ascending=False)

# Display the top 20 movies with their titles, normalized averages, popularity, and score
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame by the score column in descending order
scored_df = movies_cleaned_df.sort_values(by='score', ascending=False)

# Create a figure with specified size
plt.figure(figsize=(16, 6))

# Create a vertical bar plot for the top 10 movies based on the score
ax = sns.barplot(x=scored_df['original_title'].head(10), 
                 y=scored_df['score'].head(10), 
                 data=scored_df, 
                 palette='deep')

# Set plot title and labels
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Movie Title', weight='bold')
plt.ylabel('Score', weight='bold')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Save the plot as an image file
plt.savefig('scored_movies_vertical.png')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data for the heatmap
top_movies = scored_df[['original_title', 'score']].head(10)
pivot_df = top_movies.pivot_table(index='original_title', values='score', aggfunc='mean')

# Create a figure with specified size
plt.figure(figsize=(10, 8))

# Create a heatmap
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

# Set plot title and labels
plt.title('Best Rated & Most Popular Blend (Heatmap)', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

# Save the plot as an image file
plt.savefig('scored_movies_heatmap.png')
```
